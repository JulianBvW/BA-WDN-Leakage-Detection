# Image Generation!

## Setup

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
```

## Main

```python
BAR_WIDTH = .2
```

### Ad-Hoc

```python
NAMES_CLS = ['kNN', 'MLP']
best_adhoc_cls_accuracy    = [1.00, 1.00]
best_adhoc_cls_recall      = [1.00, 1.00]
best_adhoc_cls_specificity = [1.00, 1.00]
best_adhoc_cls_precision   = [1.00, 1.00]
NAMES_REGR = ['LR', 'Lasso', 'Ridge', 'kNN', 'MLP']
best_adhoc_regr_accuracy    = [.914, .970, .886, 1.00, 1.00]
best_adhoc_regr_recall      = [.847, .944, .771, 1.00, 1.00]
best_adhoc_regr_specificity = [.980, .994, 1.00, 1.00, 1.00]
best_adhoc_regr_precision   = [.978, .992, 1.00, 1.00, 1.00]
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4), sharey=True, gridspec_kw={'width_ratios': [len(NAMES_CLS), len(NAMES_REGR)]})

x = np.arange(len(NAMES_CLS))
ax1.bar(x-BAR_WIDTH/2-BAR_WIDTH, best_adhoc_cls_accuracy,  BAR_WIDTH)
ax1.bar(x-BAR_WIDTH/2,       best_adhoc_cls_recall,        BAR_WIDTH)
ax1.bar(x+BAR_WIDTH/2,       best_adhoc_cls_specificity,   BAR_WIDTH)
ax1.bar(x+BAR_WIDTH/2+BAR_WIDTH, best_adhoc_cls_precision, BAR_WIDTH)

plt.sca(ax1)
plt.xticks(x, NAMES_CLS)
plt.xlabel('Classification')
plt.ylabel('Score')
plt.legend(['Accuracy', 'Recall', 'Specificity', 'Precision'], loc='lower left')

x = np.arange(len(NAMES_REGR))
ax2.bar(x-BAR_WIDTH/2-BAR_WIDTH, best_adhoc_regr_accuracy,  BAR_WIDTH)
ax2.bar(x-BAR_WIDTH/2,       best_adhoc_regr_recall,        BAR_WIDTH)
ax2.bar(x+BAR_WIDTH/2,       best_adhoc_regr_specificity,   BAR_WIDTH)
ax2.bar(x+BAR_WIDTH/2+BAR_WIDTH, best_adhoc_regr_precision, BAR_WIDTH)

plt.sca(ax2)
plt.xticks(x, NAMES_REGR)
plt.xlabel('Regression-Ensemble')

f.tight_layout()
f.show()
```
